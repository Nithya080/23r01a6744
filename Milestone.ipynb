```python
import requests
import json
```

```python
url="https://raw.githubusercontent.com/mitre/cti/master/enterprise-attack/enterprise-attack.json"
data=requests.get(url)
data=data.json()
```

```python
print(data['objects'][0]['type'])
```

```python
techniques=[]
for i in range(len(data['objects'])):
    if data['objects'][i]['type']=='attack-pattern':
      techniques.append(data['objects'][i])
```

```python
print(len(techniques))
```

```python
def threat_score(tech):
  d={}
  for i in tech:
    score=5
    if("credential" in i['name'].lower()):
      score+=3
    elif("execution" in i['name'].lower()):
      score+=2
    elif("privilege" in i['name'].lower()):
      score+=3
    elif("persistence" in i['name'].lower()):
      score+=2
    elif("lateral" in i['name'].lower()):
      score+=2
    if("administrator" in i['description'].lower()):
      score+=2
    elif("remote" in i['description'].lower()):
      score+=2
    elif("bypass" in i['description'].lower()):
      score+=2
    elif('stealth' in i['description'].lower()):
      score+=1
    d[i['name']]=score
    print(i['name'],"-->",score)
  return d
d=threat_score(techniques)
```

```python
#The scoring system helps cybersecurity analysts identify, rank, and prioritize the most critical threats quickly.
```

```python
print(d)
```

```python
c=max(d.values())
cnt=0
while(cnt != 10):
  for k,v in d.items():
    if(v==c):
      print(k)
      cnt+=1
    if(cnt == 10):
      break
  c-=1
```
